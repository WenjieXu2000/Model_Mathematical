## 提取年度数据

```python
import os
files_1 = os.listdir("ProblemCData\CO2 Emission")
print(files_1)
def preprocess(file):
    data = pd.read_excel(file).iloc[2:-5,:-2]
    # 重设列名
    data_index = np.array('State')
    data_year = np.arange(1980,2017)
    column = np.hstack((data_index,data_year))
    data.columns = column
    # 重置索引
    data = data.reset_index(drop=True)
    return data
```

```python
column=['1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989']
coal_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[0])).drop(columns=column)
commercial_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[1])).drop(columns=column)
electricity_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[2])).drop(columns=column)
industrial_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[3])).drop(columns=column)
natural_gas_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[4])).drop(columns=column)
petroleum_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[5])).drop(columns=column)
residential_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[6])).drop(columns=column)
transportation_state = preprocess(os.path.join('ProblemCData\CO2 Emission',files_1[7])).drop(columns=column)
transportation_state.head()
```

## 计算能源消耗量

```python
# Coal: 1827.04 short ton 短吨--> 1kg CO2
# Petroleum: 14.91 gallon 加仑 --> 1kg CO2
# Natural gas: 54870 cubic feet 立方英尺 --> 1kg CO2
# CO2单位：百万公吨
coal_state.iloc[:, 1:] = coal_state.iloc[:, 1:] * 1827.04
natural_gas_state.iloc[:, 1:] = natural_gas_state.iloc[:, 1:] * 14.91
petroleum_state.iloc[:, 1:] = petroleum_state.iloc[:, 1:] * 54870
#得到每种能源的每年消耗量
coal_state.head()
```

# 建模

```python
import statsmodels.api as sm

params_list = []
for i in range(51):
    # 总排放量
    target = co2.iloc[i, 1:].to_numpy().reshape(-1, 1)
    # 人均GDP
    gdp = gdp_per.iloc[i, :].to_numpy().reshape(-1, 1)
    # 能源消耗量
    a = coal_state.iloc[i, 1:].to_numpy().reshape(-1, 1)
    b = natural_gas_state.iloc[i, 1:].to_numpy().reshape(-1, 1)
    c = petroleum_state.iloc[i, 1:].to_numpy().reshape(-1, 1)
    # 三种产业占比
    s1 = mining.iloc[i, 1:].to_numpy().reshape(-1, 1)
    s2 = construction.iloc[i, 1:].to_numpy().reshape(-1, 1)
    s3 = manufacturing.iloc[i, 1:].to_numpy().reshape(-1, 1)
    # 合并数据
    labels = ['GDP', 'coal', 'natural_gas', ' petroleum', 'mining', 'construction',' manufacturing']
    indexs = np.arange(1990, 2017)
    data = np.hstack((gdp,a,b,c,s1,s2,s3))
    data = pd.DataFrame(data,columns=labels,index=indexs)
    # 替换所有0值
    data[data==0] =0.01

    # 数据对数化
    data = np.log(data.astype(np.float32))
    target = np.log(target.astype(np.float32))
    X = data.to_numpy()
    Y = target 
    X = sm.add_constant(X) # 添加截距项
    model = sm.OLS(Y,X).fit()
    params = model.params
    params_list.append(params.tolist())
```

```python
labels = ['截距','GDP', 'coal', 'natural_gas', 'petroleum', 'mining', 'construction','manufacturing']
state_name = summary['State']
params = pd.DataFrame(np.array(params_list),columns=labels,index=state_name)
params.head()
```

```python
params.to_csv('ProblemCData/拟合/回归系数_2060.csv')
```

```python
fuel = pd.read_excel('ProblemCData/fuels_2016.xlsx').iloc[3:-2, :]
column = [
    'State', 'Coal', 'Petroleum', 'Natural gas', 'Total', 'Coal',
    'Shares Petroleum', 'Natural gas'
]
fuel.columns = column
# 重置索引
fuel = fuel.reset_index(drop=True)
print(fuel.shape)
# 选择排名前十和倒十的所有州
fuel_state_up = fuel.sort_values(by='Total', ascending=False).iloc[:, [0,5,6,7]]
fuel_state_up.head()
```

```python
def count_num(t):
    count = [0,0,0,0]
    for i in range(fuel_state_up.shape[0]):
        if fuel_state_up.iloc[i,t] >= 0.5:
            count[0] +=1
        if 0.4<= fuel_state_up.iloc[i,t] < 0.5:
            count[1] +=1
        if 0.3<= fuel_state_up.iloc[i,t] < 0.4:
            count[2] +=1
        if fuel_state_up.iloc[i,t] < 0.3:
            count[3] +=1
    return count

# Coal
print(count_num(1))
# 石油
print(count_num(2))
# 天然气
print(count_num(3))
```

```python
from pyecharts import options as opts
from pyecharts.charts import PictorialBar
from pyecharts.globals import SymbolType

location = [">50%", "50%~40%", "40%~30%", "30%<"]

fig = PictorialBar(
    init_opts=opts.InitOpts(width='1000px', height='500px', bg_color='white'))
fig.add_xaxis(location)
fig.add_yaxis(
    series_name='Coal',  # 系列名称，用于 tooltip 的显示，legend 的图例筛选
    y_axis=count_num(1),  # 系列数据
    symbol_size=18,  # 图形的大小
    symbol_repeat="fixed",  # 指定图形元素是否重复
    symbol_offset=[0, 0],  # 图形相对于原本位置的偏
    is_symbol_clip=True,  # 是否剪裁图形。
    symbol=SymbolType.ROUND_RECT,
    color='#8ECFC9',
    label_opts=opts.LabelOpts(is_show=False),  # # 标签配置项
)
fig.add_yaxis(
    series_name='Petroleum',  # 系列名称，用于 tooltip 的显示，legend 的图例筛选
    y_axis=count_num(2),  # 系列数据
    symbol_size=18,  # 图形的大小
    symbol_repeat="fixed",  # 指定图形元素是否重复
    symbol_offset=[0, 25],  # 图形相对于原本位置的偏
    is_symbol_clip=True,  # 是否剪裁图形。
    symbol=SymbolType.ROUND_RECT,
    color='#FA7F6F',
    label_opts=opts.LabelOpts(is_show=False),  # # 标签配置项
)
fig.add_yaxis(
    series_name='Nature Gas',  # 系列名称，用于 tooltip 的显示，legend 的图例筛选
    y_axis=count_num(3),  # 系列数据
    symbol_size=18,  # 图形的大小
    symbol_repeat="fixed",  # 指定图形元素是否重复
    symbol_offset=[0, -25],  # 图形相对于原本位置的偏
    is_symbol_clip=True,  # 是否剪裁图形。
    symbol=SymbolType.ROUND_RECT,
    color='#BEB8DC',
    label_opts=opts.LabelOpts(is_show=False),  # # 标签配置项
)

fig.reversal_axis()
fig.set_global_opts(
    xaxis_opts=opts.AxisOpts(is_show=False),
    yaxis_opts=opts.AxisOpts(
        axislabel_opts=opts.LabelOpts(font_size=14),# 修改字体大小
        axistick_opts=opts.AxisTickOpts(is_show=False),
        axisline_opts=opts.AxisLineOpts(linestyle_opts=opts.LineStyleOpts(
            opacity=0)),
    ),
    legend_opts=opts.LegendOpts(is_show=True, pos_top=35,),
)

fig.render_notebook()
```
