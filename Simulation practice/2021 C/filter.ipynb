```python
data_path1 = r"C:\Git Code\Mathematical-modeling\Simulation practice\2021 C\Data\DataSet.xlsx"
dataset_1 = pd.read_excel(data_path1, engine='openpyxl',header=0, index_col=False)
dataset_1.head()
```

```python
data_list = []
for i in dataset_1.index:
    if dataset_1.iloc[i]["Lab Status"] == "Unprocessed" or dataset_1.iloc[i]["Lab Status"] == "Unverified":
        continue
    if str(dataset_1.iloc[i]["Detection Date"])[:4] != "2019" and str(dataset_1.iloc[i]["Detection Date"])[:4] != "2020":
        print(str(dataset_1.iloc[i]["Detection Date"])[:4])
        continue
    data_list.append(i)
dataset_1 = dataset_1.iloc[data_list, [0,1,3,5,6,7]]
dataset_1.to_csv(r"C:\Git Code\Mathematical-modeling\Simulation practice\2021 C\Data\DataSet.csv", index=False, header=True)
```
